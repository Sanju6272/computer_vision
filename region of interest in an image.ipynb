```python
import cv2 as cv
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv.imread("range_rover_sport.jpg")
cv.imshow("image",img)

#in order to plot in graph we need to change the bgr image to rgb image
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(img_rgb)
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```
